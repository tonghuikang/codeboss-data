```python
import numpy as np
import pandas as pd
df = pd.read_csv("./phone_list.csv")
df.head()
```

Objective - Data cleansing of SG phone number	
1	The [Phone List] contain 1000 customer contact no, with mixture of Singapore and foreign country phone number.
2	"To identify and reformat the Singapore phone number:
- The formatted number should be 8-digit with prefix of [3, 6, 8, 9]
- For identified Singapore phone number, set the Country Code to ""SG - Singapore""
"
3	"In [Findings and Solutions], provide 5 findings and proposed solution.
"
4	"To cleanse the phone number and output the result in [Cleansed] tab.
"

```python
def check_valid(string):
    if len(string) > 0:
        if string[0] == "3" or string[0] == "6" or string[0] == "8" or string[0] == "9":
            if len(string) == 8:
                return True
    return False

def convert_hex_if_got_letter(txt):
    # preliminary hex key conversion if contain any letter
    try:
        if txt.upper().isupper():
            txt = str(int(txt, 16))
            df.loc[i, "Contain Hex"] = True
    except:
        return txt 
    return txt
    

df["Valid"] = False
df["Contain Hex"] = False
df["Updated Country"] = False
df["From Multiple"] = False
df["Cleansed Country Code"] = df["Country Code"] 
df["Cleansed Phone Number"] = ""
df_copy = df.copy()
```

```python
rows_with_duplicates = []

for i,txt in enumerate(df["Phone Number"]):
    # basic universal cleansing
    txt = str(txt)
    if txt == "63392801 TO 04":
        txt = "63392801;63392802;63392803;63392804"
    omit = (' ', '.', '-')
    for r in omit:
        txt = txt.replace(r, "")

    if txt == "nan":
        txt = ""
        
    replacements = (',', ';', '/')
    for r in replacements:
        txt = txt.replace(r, '$$$$$')
    if '$$$$$' in txt:
        rows_with_duplicates.append(i)
        for tx in txt.split('$$$$$'):
            row = df.loc[i].copy()
            row["From Multiple"] = True
            row["Phone Number"] = tx
            df_copy = df_copy.append(row, ignore_index=True)
        else:
            continue
            
df_copy = df_copy.reset_index(drop=True)
df = df_copy.drop("ID", axis=1)
del df_copy
df.tail(100)
```

```python
for i,txt in enumerate(df["Phone Number"]):
    # basic universal cleansing
    txt = str(txt).replace(" ","").replace("-", "").replace(".", "")
    txt = txt.lower()
    if txt == "nan":
        txt = ""

    # preliminary hex key conversion if contain any letter
    txt = convert_hex_if_got_letter(txt)        
        
    # check validity
    valid = check_valid(txt)
    if valid:
        df.loc[i, "Valid"] = True
        df.loc[i, "Cleansed Phone Number"] = txt
        
        if str(df.loc[i, "Cleansed Country Code"]) == "nan":
            df.loc[i, "Cleansed Country Code"] = "SG - Singapore"
            df.loc[i, "Updated Country"] = True
        continue
    
    # Cleansed Country Code
    if "+65" in txt or "+65" in txt or "(65)" in txt:
        df.loc[i, "Cleansed Country Code"] = "SG - Singapore"
        df.loc[i, "Updated Country"] = True
        
        try:
            txt = txt.replace("(+65)","").replace("(65)","").replace("+65","")
            
            txt = convert_hex_if_got_letter(txt)
            
            if valid:
                df.loc[i, "Valid"] = True
                df.loc[i, "Cleansed Phone Number"] = txt
                continue
            
            txt = txt[-8:]
            valid = check_valid(txt)
            if valid:
                df.loc[i, "Valid"] = True
                df.loc[i, "Cleansed Phone Number"] = txt
                continue
        except:
            pass
        
    df.loc[i, "Cleansed Phone Number"] = txt
```

```python
df = df.drop(df.index[rows_with_duplicates])
pd.set_option('max_rows', 1000)
df.head(1000)
```

```python
df.to_csv("cleansed.csv")
```
